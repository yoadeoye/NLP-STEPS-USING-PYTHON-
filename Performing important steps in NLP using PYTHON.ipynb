### Performing important steps in NLP using PYTHON 

### Introduction

```python
# FIRST I IMPORT SPACY AS IT IS THE NLTK I WANT TO USE 
import spacy
nlp=spacy.load('en_core_web_lg')
project_doc= open("punchNewsToday.txt",encoding='UTF-8').read()
project_doc=nlp(project_doc)
print(project_doc)
```

##### Senticizing: 
I refer this to the process of converting
my text documents into sentences it contained therein

```python
#setence detection
sentence=list(project_doc.sents)
print(sentence)
len(sentence) # 18 list of sentences with the text
```

##### Tokenization:
This is majorly splitting the text data 
into elements such as words,punctuactions it contained

```python
# tokenization
new_token=[]
for token in project_doc:
    new_token.append(token)
    print(token.text)
new_token
len(project_doc)  # 392 tokens : words,punctuations,space present 
len(new_token)==len(project_doc)  # getting to confirm the no of tokens
```

##### Part of Speech and Dependency
Identifying the part of speech and its dependency to know the class it belongs to

```python
## detect part of speech and dependency
for token in project_doc:
    print(token.text,  "  PART OF SPEECH :  ",token.pos_,"   DEPENDENCY  :  ",token.dep_)
```

##### Stop Word and Punctuation
Stop Word are words that are not necessarily adding meaning to a context

```python
## checking for stop words and punctuations
for token in project_doc:
    print(token.text, "    STOP_WORD   :  ",token.is_stop, "  PUNCTUATION :   ", token.is_punct)
```

##### Lemmatization
A process of identifying the root(either prefix or suffix) of particular word.Lemmatization still make sense of the word even if the lemma is removed unlike stemming 

```python
#lemmatization
for token in project_doc:
    print(token.text, "  LEMMA   : ",token.lemma_)
```

#### Named Entity recognition 
Identifying the group or collection to which a particular word belongs

```python
# Showing the entity to which each word belongs and their position 
for ent in project_doc.ents:
    print(ent.text,ent.label_,ent.start_char,ent.end_char)    
```

###### WORD VECTOR/EMBEDDING&SIMILARITY
This is the process of mapping words or phrases 
from vocabulary to a corresponding vector of real numbers
to find predictions or word similaritie i.e representing words as a vector.

```python
## WORD VECTOR/EMBEDDING&SIMILARITY
import numpy as np
import pandas as pd
import seaborn as sns
data=[]
for d in range(18):
    for i in range(len(sentence)-1):
        data.append(np.round(sentence[d].similarity(sentence[i]),2))
data=pd.DataFrame(np.array(data).reshape(18,17))
data
```
